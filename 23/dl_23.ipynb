```python
X_train = []
for sentence in x_train:
    idx_sentence = []
    for word in sentence.split():
        idx_sentence.append(word_to_idx[word.lower()])
    if len(idx_sentence)<max_length_sentence:
        idx_sentence = idx_sentence+(max_length_sentence-len(idx_sentence))*[0]
    X_train.append(idx_sentence)
X_train = np.array(X_train)
Y_train = np.array(np.eye(5)[y_train], dtype='float32')

X_test = []
for sentence in x_test:
    idx_sentence = []
    for word in sentence.split():
        idx_sentence.append(word_to_idx[word.lower()])
    if len(idx_sentence)<max_length_sentence:
        idx_sentence = idx_sentence+(max_length_sentence-len(idx_sentence))*[0]
    X_test.append(idx_sentence)
X_test = np.array(X_test)
Y_test = np.array(np.eye(5)[y_test], dtype='float32')
```

```python
embedding_matrix = np.array(list(word_to_vec.values()))
embedding_matrix.shape
```

```python
def Model(params):
    dims = params['dims']
    embedding_matrix = params['embedding_matrix']
    X_train = params['X_train']
    Y_train = params['Y_train']
    epochs = params['epochs']
    learning_rate = params['learning_rate']
    batch_size = params['batch_size']
    dropout_rate = params['dropout_rate']
    embedding_matrix = params['embedding_matrix']
    trainable = params['trainable']
    f1 = tf.keras.layers.Embedding(input_dim=embedding_matrix.shape[0], 
                                   output_dim=embedding_matrix.shape[1], 
                                   name='embedding')
    f2 = tf.keras.layers.LSTM(units=128,
                              activation='tanh',
                              recurrent_activation='sigmoid',
                              kernel_initializer='glorot_uniform',
                              bias_initializer='zeros',
                              recurrent_initializer='zeros',
                              return_sequences=True,
                              return_state = False)
    f3 = tf.keras.layers.Dropout(rate=dropout_rate)
    f4 = tf.keras.layers.LSTM(units=128,
                              activation='tanh',
                              recurrent_activation='sigmoid',
                              kernel_initializer='glorot_uniform',
                              bias_initializer='zeros',
                              recurrent_initializer='zeros',
                              return_sequences=False,
                              return_state = False)
    f5 = tf.keras.layers.Dropout(rate=dropout_rate)
    f6 = tf.keras.layers.Dense(units=dims[1], 
                              activation='softmax', 
                              kernel_initializer='glorot_uniform', 
                              bias_initializer='zeros')
    x = tf.keras.Input(shape=dims[0])
    a1 = f1(x)
    a2 = f2(a1)
    a3 = f3(a2)
    a4 = f4(a3)
    a5 = f5(a4)
    y = f6(a5)
    model = tf.keras.Model(x, y)   
    model.get_layer(name='embedding').trainable = trainable
    model.get_layer(name='embedding').set_weights([embedding_matrix])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate, 
                                         beta_1=0.9, 
                                         beta_2=0.999, 
                                         epsilon=1e-07)
    model.compile(loss='categorical_crossentropy', metrics=['categorical_accuracy'], optimizer=optimizer)
    model.summary()
    model.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size)
    return model
```

```python
model = Model({'dims': [X_train.shape[1], Y_train.shape[1]], 
               'X_train': X_train,
               'Y_train': Y_train,
               'epochs': 50, 
               'learning_rate': 0.01,
               'batch_size': 32,
               'dropout_rate': 0.5,
               'embedding_matrix': embedding_matrix, 
               'trainable': False})
```

```python
model.evaluate(X_train, Y_train)
```

```python
model.evaluate(X_test, Y_test)
```

```python
Y_test_pred = model.predict(X_test)

for idx in range(10):
    emoji = label_to_emoji(np.argmax(Y_test[idx,:]))
    emoji_prediction = label_to_emoji(np.argmax(Y_test_pred[idx,:]))
    print('Sentence: {}'.format(x_test[idx]))
    print('Original Emoji: {}'.format(emoji))
    print('Predicted Emoji: {}'.format(emoji_prediction))
    print()
```
